```python
import pandas as pd
import re
```

```python
def create_df():
    fo = open("./result.json", "r")
    Dict = {}
    titles,price,commits,total_num = [],[],[],[]
    for line in fo.readlines():
        if len(line) >2:
            text = line[line.find("{")+1:line.find("}")]
            text_ = text.split("\"")
            if len(text_)>15:
                titles.append(text_[3])
                price_i = re.findall(r"\d+\.?\d*",str(text_[7]))
                #print(price_i)
                if not price_i:
                    price.append("None")
                else:
                    price.append(re.findall(r"\d+\.?\d*",str(price_i))[0])
                commits.append(text_[11].split(" ")[0])
                total_num.append(text_[15].replace(',', ''))
    Dict["title"] = titles
    Dict["price"] = price
    Dict["commits"] = commits
    Dict["total_num"] = total_num
    df = pd.DataFrame(Dict)
    return df
```

```python
df = create_df()
df
```

```python
df.info()
```

Transform the object to num

```python
df = df[-df.price.isin(["None"])]
df.info()
```

```python
df["price"] = df["price"].astype(float)
df["commits"] = df["commits"].astype(float)
df["total_num"] = df["total_num"].astype(int)
```
